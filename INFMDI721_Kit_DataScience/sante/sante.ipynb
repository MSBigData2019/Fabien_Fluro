```python
df_medecin = pd.read_excel('/Users/fabienfluro/Documents/MS_BGD/INFMDI721_Kit_DataScience/Repo_Git/Fabien_Fluro/sante/rpps-medecins18-tab2v5_71029108888648.xlsx', sheet_name='clean', index_col = [0], header=[0], skipinitialspace=True)

df_medecin["REGION"] = df_medecin.index.get_level_values('ZONE INSCRIPT')

df_medecin = df_medecin.reset_index()

df_medecin = df_medecin.drop(['ZONE INSCRIPT', 'Moins de 30 ans', 'Entre 30 et 34 ans', 'Entre 35 et 39 ans', 'Entre 40 et 44 ans', 'Entre 45 et 49 ans', 'Entre 50 et 54 ans', 'Entre 55 et 59 ans', 'Entre 55 et 59 ans', 'Entre 60 et 64 ans', 'Entre 65 et 69 ans', '70 ans et plus'], axis=1)

# Change columns name
df_medecin.columns = ['SPECIALITE', 'NUMBER', 'REGION']


```

```python
#check regions , we keep only department
df_medecin['REGION'].unique()
```

```python
# Garder les regions qui comporte un numero de département
df_medecin = df_medecin[df_medecin['REGION'].str.contains(r'^[0-9]{2,3}.*')]

#reg = r'([\D]*)(?P<Dosage>\d+)(.*),(?P<Galenique>.*)'

# Split numero et nom du département
reg = r'(?P<NUM_DEP>[0-9]{2,3}) - (?P<NOM_DEP>.*)'
df_medecin = pd.concat([df_medecin, df_medecin['REGION'].str.extract(reg, expand=True)], axis=1)


# re order
df_medecin = df_medecin[['NUM_DEP', 'NOM_DEP', 'SPECIALITE', 'NUMBER']]
```

```python
df_medecin = df_medecin[df_medecin['SPECIALITE'] == 'SpÈcialistes']
df_medecin.head(5)
```
